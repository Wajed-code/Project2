```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.DataFrame({
  "Department": ["Sales", "Marketing", "IT", "HR", "Finance", "Operations", "Logistics", "R&D", "Legal", "Admin"],
  "Revenue": [500000, 350000, np.nan, 150000, 400000, 450000, 300000, -10000, "", 250000], 
  "Expenses": [200000, np.nan, 100000, 80000, 150000, "", 180000, 90000, 120000, 220000], 
  "Profit": [300000, 170000, 100000, 70000, 250000, 200000, np.nan, -20000, 130000, 30000], 
  "Budget Allocation": [550000, 370000, 220000, "", 420000, 460000, 320000, 0, 210000, np.nan], 
  "Variance": [50000, 20000, np.nan, 30000, 20000, 10000, -5000, 15000, 0, 8000], 
  "Financial Risk": ["Low", "Medium", "Low", "High", "", "Medium", "High", "Low", "Medium", "High"] 
})
data 
```

```python
data.replace("", np.nan, inplace=True)
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#      
missing_values = data.isnull().sum()

#  Barplot   
plt.figure(figsize=(10, 6))
sns.barplot(x=missing_values.index, y=missing_values.values, palette="coolwarm")
plt.title("Missing Values per Column")
plt.xlabel("Columns")
plt.ylabel("Missing Values Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6), facecolor='#F8F8F8')   
sns.scatterplot(x=range(len(data)), y=data['Revenue'], color="#0000FF")   
plt.title("Revenue Values")
plt.xlabel("Index")
plt.ylabel("Revenue")
plt.grid(True, linestyle='--')   
plt.show()
```

```python
plt.figure(figsize=(8, 6), facecolor='#F8F8F8')
sns.boxplot(x=data['Profit'])
plt.title("Boxplot for Outlier Detection")
plt.xticks(rotation=45)
plt.show()
```

```python
data = data.dropna(subset=['Department', 'Revenue', 'Expenses', 'Profit','Financial Risk',]) # 
data
```

```python
data['Profit'] = data['Profit'].apply(lambda x: x if x >= 0 else np.nan) #    
data
```

```python
data.loc[:,'Profit'].fillna('Unknown', inplace=True)
data #   
```

```python
MV = data["Budget Allocation"].mean(skipna=True)
data["Budget Allocation"] = data["Budget Allocation"].apply(lambda x: x if pd.notna(x) and x > 0 else MV)
data["Budget Allocation"] = data["Budget Allocation"].round(1)
data # 
```

```python
data['Variance'] = data['Variance'].apply(lambda x: x if x >= 0 else np.nan)
data
```

```python
FR = {'Low': 0, 'Medium': 1, 'High': 2}
data['Financial Risk'] = data['Financial Risk'].map(FR) 
data
```

```python
data['Financial Risk'].fillna('Unknown', inplace=True)
data
```

```python
data = data.drop_duplicates()
data
```

```python
data.replace("Unknown", 0, inplace=True)
data
```
