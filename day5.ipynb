```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Revenue': [50000, 52000, 53000, 55000, 56000, 62000, 58000, 57000, 59000, 60000, 59000, 61000],
    'Expenses': [30000, 32000, 34000, 35000, 36000, 34000, 32000, 31000, 30000, 31000, 32000, 33000],
    'Profit': [20000, 22000, 19000, 20000, 22000, 28000, 26000, 26000, 29000, 29000, 27000, 28000],
    'Marketing': [15000, 15500, 16000, 16500, 17000, 17500, 16900, 16500, 16000, 15800, 15700, 16000]
})

plt.figure(figsize=(7, 5))
plt.plot(data['Month'], data['Revenue'], marker='o', linestyle='--', label='Revenue', color='r')
plt.plot(data['Month'], data['Expenses'], marker='s', linestyle='--', label='Expenses', color='g')
plt.plot(data['Month'], data['Profit'], marker='^', linestyle='--', label='Profit', color='b')
plt.xlabel('Month')
plt.ylabel('Amount ($)')
plt.title('Revenue, Expenses, and Profit Trend Over the Year')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
print(data.head())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Revenue': [50000, 52000, 48000, 51000, 53000, 55000, 60000, 62000, 58000, 57000, 59000, 61000],
    'Expenses': [30000, 32000, 31000, 30000, 34000, 33000, 35000, 36000, 34000, 32000, 31000, 30000],
    'Profit': [20000, 20000, 17000, 21000, 19000, 22000, 25000, 26000, 24000, 25000, 28000, 31000],
})

month = 'Jan'
values = [data.loc[data['Month'] == month, 'Revenue'].values[0],
          data.loc[data['Month'] == month, 'Expenses'].values[0],
          data.loc[data['Month'] == month, 'Profit'].values[0]]
labels = ['Revenue', 'Expenses', 'Profit']

plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title(f'Distribution of Revenue, Expenses, and Profit for {month}')
plt.axis('equal')  
plt.show()
print(data.head())
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Revenue': [50000, 52000, 48000, 51000, 53000, 55000, 60000, 62000, 58000, 57000, 59000, 61000],
    'Expenses': [30000, 32000, 31000, 30000, 34000, 33000, 35000, 36000, 34000, 32000, 31000, 30000],
    'Profit': [20000, 20000, 17000, 21000, 19000, 22000, 25000, 26000, 24000, 25000, 28000, 31000],
})

data_for_heatmap = data.set_index('Month').T

plt.figure(figsize=(10, 6))
sns.heatmap(data_for_heatmap, annot=True, cmap='coolwarm', cbar=True)
plt.title('Heatmap of Revenue, Expenses, and Profit Over the Year')
plt.show()
print(data.head())
```
